```python
import pygame
import random

# 初始化 Pygame
pygame.init()

# 定义游戏窗口的尺寸
WIDTH, HEIGHT = 800, 600
win = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Square and Ball Game")

# 定义颜色
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)

# 定义正方形的属性
SQUARE_SIZE = 20
square_x = WIDTH // 2 - SQUARE_SIZE // 2
square_y = HEIGHT // 2 - SQUARE_SIZE // 2
square_dx = 0
square_dy = 0

# 获取用户输入的起始方向
while True:
    try:
        direction = int(input("请输入起始方向 (1: 上, 2: 下, 3: 左, 4: 右): "))
        if direction == 1:
            square_dy = -5
        elif direction == 2:
            square_dy = 5
        elif direction == 3:
            square_dx = -5
        elif direction == 4:
            square_dx = 5
        else:
            print("输入无效，请输入 1 - 4 之间的数字。")
            continue
        break
    except ValueError:
        print("输入无效，请输入一个整数。")

# 定义小球的属性
BALL_SIZE = 50

balls = []
ball = {
    "x": random.randint(0, WIDTH - BALL_SIZE),
    "y": random.randint(0, HEIGHT - BALL_SIZE),
    "dx": random.choice([-3, 3]),
    "dy": random.choice([-3, 3])
}
balls.append(ball)

clock = pygame.time.Clock()

running = True
while running:
    clock.tick(60)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # 移动正方形
    square_x += square_dx
    square_y += square_dy

    # 正方形边界检测和反弹
    if square_x < 0 or square_x > WIDTH - SQUARE_SIZE:
        square_dx = -square_dx
    if square_y < 0 or square_y > HEIGHT - SQUARE_SIZE:
        square_dy = -square_dy

    # 移动小球
    for ball in balls[:]:
        ball["x"] += ball["dx"]
        ball["y"] += ball["dy"]

        # 小球边界检测和反弹
        if ball["x"] < 0 or ball["x"] > WIDTH - BALL_SIZE:
            ball["dx"] = -ball["dx"]
        if ball["y"] < 0 or ball["y"] > HEIGHT - BALL_SIZE:
            ball["dy"] = -ball["dy"]

        # 检测正方形和小球的碰撞
        if (
            square_x < ball["x"] + BALL_SIZE
            and square_x + SQUARE_SIZE > ball["x"]
            and square_y < ball["y"] + BALL_SIZE
            and square_y + SQUARE_SIZE > ball["y"]
        ):
            balls.remove(ball)
            # 生成三个新的小球
            for _ in range(3):
                new_ball = {
                    "x": WIDTH // 2 - BALL_SIZE // 2,
                    "y": HEIGHT // 2 - BALL_SIZE // 2,
                    "dx": random.choice([-3, 3]),
                    "dy": random.choice([-3, 3])
                }
                balls.append(new_ball)

    # 检查小球数量是否达到一定数值
    if len(balls) >= 100000:
        running = False

    # 绘制背景
    win.fill(WHITE)

    # 绘制正方形
    pygame.draw.rect(win, RED, (square_x, square_y, SQUARE_SIZE, SQUARE_SIZE))

    # 绘制小球
    for ball in balls:
        pygame.draw.circle(win, BLACK, (ball["x"], ball["y"]), BALL_SIZE)

    pygame.display.update()

pygame.quit()
```
